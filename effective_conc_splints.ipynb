Assumes midpoint method:

```python
print('What is the length of the template strand?')
n_nt = int(input())
max_e2e_dist = 0.34*n_nt    # maximal base pair distance in nm
print('The maximal distance that the base pairs can have is approximately ' + str(max_e2e_dist) + ' nm.')
```

```python
occupied_vol = (max_e2e_dist/(1e9))**3    # in cubic meters
print('The volume one molecule occupies is ' + str(occupied_vol) + ' cubic meters.')
```

```python
molecules_per_cubic_m = 1/occupied_vol
print('There are ' + str(molecules_per_cubic_m) + ' molecules per cubic meters.')
Avogadro_num = 6.02214076e23
eff_conc = molecules_per_cubic_m*(1/Avogadro_num)*(1/(10**3))    # measured in molars
print('The effective concentration is ' + str(eff_conc) + ' M = ' + str(1e6*eff_conc) + ' uM')
```

```python
import matplotlib.pyplot as plt

Avogadro_num = 6.02214076e23
eff_conc_list = []
x = range(41, 72)
for n_nt in x:
    max_e2e_dist = 0.34*n_nt    # maximal base pair distance in nm
    occupied_vol = (max_e2e_dist/(1e9))**3    # in cubic meters
    molecules_per_cubic_m = 1/occupied_vol
    eff_conc = 1e6*(molecules_per_cubic_m*(1/Avogadro_num)*(1/(10**3)))    # measured in micro molars
    eff_conc_list.append(eff_conc)

plt.plot(x, eff_conc_list, '.-')
plt.xlabel('Template strand length (nt)')
plt.ylabel('Effective concentration ($\mu$M)')
plt.savefig('eff_conc_vs_template_len.png', dpi=600, bbox_inches="tight")
```
